# Cálculo de la matriz de correlación de Pearson

Este notebook muestra cómo calcular y visualizar la matriz de correlación de Pearson usando el archivo `cleaned_dataset.csv` (una versión limpia del Pima Indian Dataset).

```python
# Importar librerías necesarias
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Cargar el dataset

Leemos el archivo `cleaned_dataset.csv` en un DataFrame de pandas.

```python
# Cargar el dataset limpio
df = pd.read_csv('../../datasets/pima_indian_diabetes_dataset/cleaned_dataset.csv')
df.head()
```

## Calcular la matriz de correlación de Pearson

Utilizamos el método `.corr()` de pandas para obtener la matriz de correlación.

```python
# Calcular la matriz de correlación de Pearson
correlation_matrix = df.corr(method='pearson')
correlation_matrix
```

## Visualizar la matriz de correlación

Usamos un heatmap de seaborn para visualizar la matriz de correlación.

```python
# Visualizar la matriz de correlación con un heatmap azul-blanco-rojo
plt.figure(figsize=(10,8))
sns.heatmap(
    correlation_matrix, 
    annot=True, 
    cmap='bwr',  # Blue-White-Red
    center=0,
    fmt=".2f"
)
plt.title('Matriz de correlación de Pearson')
plt.show()
```
